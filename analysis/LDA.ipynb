```python
from gensim.models.ldamodel import LdaModel
from gensim.corpora import Dictionary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mlb = pd.read_csv("../data/tidied_text_added.csv")
```

```python
mlb.head()
```

```python
mlb.sort_values("tweet_length",ascending=False).head(1)
```

```python
# all_words = [word for tokens in mlb.cleaned_text_list for word in tokens]
mlb["tweet_length"]  = [len(tokens) for tokens in mlb.cleaned_text_list]
mlb_cleaned = mlb[mlb.tweet_length<=140 ]
mlb_cleaned = mlb_cleaned[mlb_cleaned.tweet_length>= 5]
tweet_length = mlb_cleaned.tweet_length
```

```python
vocab = sorted(list(set(mlb.cleaned_text_list)))

# print('{} words total, with a vocabulary size of {}'.format(
#     len(all_words), len(vocab)))
print('Max tweet length is {}'.format(max(tweet_length)))
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(tweet_length)
plt.title('Tweet Length Distribution', fontsize=18)
plt.xlabel('Words per Tweet', fontsize=12)
plt.ylabel('Number of Tweets', fontsize=12)
plt.show()
```
